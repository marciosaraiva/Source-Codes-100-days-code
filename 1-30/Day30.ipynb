**Exercise: Array Fill II**

Write a program that reads a number T and fill a vector N[1000] with the numbers from 0 to T-1 repeated times, like as the example below.

**Input:**
The input contains an integer number T (2 ≤ T ≤ 50).

3

**Output:**
For each position of the array N, print "N[i] = x", where i is the array position and x is the number stored in that position.

N[0] = 0

N[1] = 1

N[2] = 2

N[3] = 0

N[4] = 1

N[5] = 2

N[6] = 0

N[7] = 1

N[8] = 2

...

```python
T = int(input())

values_of_T = []

for v in range(T):
    values_of_T.append(v)

N = []
#for x in range(1000):
for x in range(20):
      for v in values_of_T:
            N.append(v)
            
#for x in range(1000):
for x in range(20):
    print("N[{}] = {}".format(x,N[x]))
```

**Exercise:Array Fill III**

Read a number X. Put this X at the first position of an array N [100]. In each subsequent position (1 up to 99) put half of the number inserted at the previous position, according to the example below. Print all the vector N.

**Input:**
The input contains a double precision number with four decimal places.

200.0000

**Output:**
For each position of the array N print "N[i] = Y", where i is the array position and Y is the number stored in that position. Each number of N[...] must be printed with 4 digits after the decimal point.

N[0] = 200.0000

N[1] = 100.0000

N[2] = 50.0000

N[3] = 25.0000

N[4] = 12.5000

...

```python
x = float(input())
x *=2
N = []
for position in range(100):
    N.append(x/2)
    print("N[{}] = {:.4f}".format(position, N[position]))
    x/=2
```

**Exercise: Array Fill IV**

In this problem you need to read 15 numbers and must put them into two different arrays: par if the number is even or impar if this number is odd. But  the size of each of the two arrrays is only 5 positions. So every time you fill one of two arrays, you must print the entire array to be able to use it again for the next numbers that are read. At the end, all remaining numbers of each one of these two arrays must be printed beggining with the odd array. Each array can be filled how many times are necessary.

**Input:**
The input contains 15 integer numbers.

**Output:**
Print the output like the following example.

```python
par = []
impar = []

for x in range(15):
    number = int(input())
    if number%2==0:
        if len(par)<5:
            par.append(number)
        else:
            for position in range(5):
                print("par[{}] = {}".format(position,par[position]))
            par=[]
            par.append(number)
    else:
        if len(impar)<5:
            impar.append(number)
        else:
            for position in range(5):
                print("impar[{}] = {}".format(position,impar[position]))
            impar=[]
            impar.append(number)

for position in range(len(impar)):
    print("impar[{}] = {}".format(position,impar[position]))

for position in range(len(par)):
    print("par[{}] = {}".format(position,par[position]))
```
